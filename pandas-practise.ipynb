1. fillna()

   a)df.fillna(value) #replaces null values in the col with value
   b) df.fillna(df.col.mean()) #fills the null value with the mean of the col

2. interpolate()

    df.colname.inetrpolate() #fills the missing value by observing the pattern from the neighbors

3. dropna()

    a) df.dropna(axis=0, how='any', subset=['col_name'], thresh=None, inplace=True)

```python
#shift-->find the details of day before or something..used for comparision in time series data
#shift backwards by shift(-1)
cof["yesterday_revenue"] = cof["revenue"].shift(1)
cof.head()
```

```python
#finding percentage change in revenue

cof["pct_change"] = cof["revenue"] / cof["yesterday_revenue"] *100
cof.head()
```
