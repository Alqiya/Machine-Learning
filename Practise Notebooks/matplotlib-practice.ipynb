### **Bar Charts**
compares categories of data by representing each category by a bar


```python
categories = np.array(["Grains", "Fruits", "Vegetables", "Proteins","Dairy", "Sweets"]) 
values = np.array([3, 1, 4, 2, 1,5])

# plt.bar(categories, values, color = 'skyblue')

#for horizontal barchart
plt.barh(categories, values, color = 'skyblue')
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.title("Daily Consumption")
plt.show()
```

#### **Pie Chart**
A pie chart is a circular graph used to display data as parts of a whole.
It divides a circle into sectors (or "slices") where the arc length or angle of each slice is proportional to the quantity or percentage it represents.
Its great for visualizing distribution among categories


```python
levels = ["freshman", "sophomore", "junior", "senior"]
vals = [300, 275, 250, 290]
colors = ["red", "yellow", "blue" ,"green"]

plt.pie(vals, 
        labels = levels,
        colors = colors,
        explode = [0,0,0,0.1],
        shadow = True,
        startangle = 90,
       autopct = "%1.1f%%") #autopct-->auto percentage... %1.1f(give float decimal with 1 precision)

plt.title("College")
plt.show()
```

#### **Scatter Plot**
Shows the relationship between two variables
Helps to identify a correlartion (+ve, -ve, None)
Example: study hours vs test scores

```python
# study_hours = x
# test_scores =y

x1 = [0,1,1,2,3,4,5,6,7,7,8]
y1 = [55, 60, 65, 75, 80, 88,90, 92, 95,95, 97]

x2 = [1,2,3,3,4,4,4,5,6,7,8]
y2 = [60, 61, 63, 63, 75, 88,90, 90, 95,97, 99]

plt.xlabel("Hours studied")
plt.ylabel("Scores")
plt.title("Test Scores of 11 students")
plt.scatter(x1, y1, color = 'hotpink', alpha =0.5, s =150, label = 'class-A')
plt.scatter(x2, y2, color = 'red', alpha =0.5, s=150, label = 'class-B')
#alpha-->transparency(0.5=50percent)

plt.legend()
plt.show()
```

#### **Histograms**
Its a visual representation of the ditribution of quantitative data
They group values into intervals(bins) and count how many are present in each range/bin.

Useful for statistical distributions
Usually a bell shaped curve is found(Normal/Gaussian Distrbution)

```python
scores = np.random.normal(loc = 80, scale = 12, size =100)
#.normal for normal disrubtion 
# loc for the upper range of value
#scale for deviation
# size = no of test scores

scores = np.clip(scores, 0, 100) #setting the min and max values so taht if there is a score that is generated below 0 is 0 and above 100 becomes 100

plt.figure()

plt.hist(scores, bins = 10, color ="lightgreen", edgecolor = 'black')

plt.xlabel("Scores")
plt.ylabel("No of students")
plt.title("Scores of 100 students")

plt.show()
```

#### **Subplots**
1. Figure - entire canvas
2. Ax - A single plot(subplot)

```python
figure , axes = plt.subplots(2,2)

x = np.array([1,2,3,4,5])

axes[0,0].plot(x, x*2, color = 'red')
axes[0,0].set_title("x*2")

axes[0,1].plot(x, x**2, color = 'blue')
axes[0,1].set_title("x**2")

axes[1,0].plot(x, x**3, color = 'green')
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**4, color = 'yellow')
axes[1,1].set_title("x**4")

plt.tight_layout() #some plots overlap, so to prevent that use this fucntion

plt.show()
```

```python
figure , axes = plt.subplots(2,2)

x = np.array([1,2,3,4,5])

axes[0,0].plot(x, x*2, color = 'red')
axes[0,0].set_title("x*2")

axes[0,1].bar(x, x**2, color = 'blue')
axes[0,1].set_title("x**2")

axes[1,0].scatter(x, x**3, color = 'green')
axes[1,0].set_title("x**3")

axes[1,1].barh(x, x**4, color = 'yellow')
axes[1,1].set_title("x**4")

plt.tight_layout() #some plots overlap, so to prevent that use this fucntion

plt.show()
```
